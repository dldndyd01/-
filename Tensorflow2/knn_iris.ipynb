```python
import pandas as pd
import numpy as np
```

```python
iris = pd.read_csv("data/iris.csv")
iris.head()
```

```python
iris.shape
```

```python
import matplotlib.pyplot as plt

groups = iris.groupby("name")
fig, ax = plt.subplots()
for name, group in groups:
    ax.plot(group.sepal_length,
            group.sepal_width,
            marker = "o",
            linestyle = "",
            label = name)
    
ax.legend(fontsize = 12, loc = "upper left")
plt.title("Scatter Plot of iris by matplotlib", fontsize = 20)
plt.show()
```

```python
X = iris.to_numpy()
X = X[:,:4]
X
```

```python
from sklearn.preprocessing import scale, minmax_scale
Z = scale(X)
Z

# minmax_scale로표준화하는 과정
```

```python
iris["target"] = iris["name"].map({"setosa":0, "versicolor" : 1, "virginica" :2})
y = np.array(iris["target"])
y
```

```python
Z, y
```

```python
from sklearn.model_selection import train_test_split
Z_train, Z_test, y_train, y_test = train_test_split(Z,y,test_size = 0.5)
```

```python
Z_train.shape
```

```python
from sklearn import neighbors
knn = neighbors.KNeighborsClassifier(n_neighbors = 5)
knn.fit(Z_train, y_train)
pred = knn.predict(Z_test)
pred
```

```python
y_test
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test,pred))
```
